```python
# Mobile Recharge Validation System

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        return "Recharge amount should be minimum ₹50"
    elif amount not in valid_plans:
        return "Invalid plan selected"
    else:
        return "Recharge Successful "

while True:
    amt = int(input("Enter recharge amount: "))
    result = validate_recharge(amt)
    print(result)

    if result == "Recharge Successful ":
        break
    else:
        print("Please try again...\n")
```

```python
# Inventory Reorder Alert System

def check_inventory(stock_dict):
    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(product, "-> Reorder Alert")
        else:
            print(product, "-> Stock OK")

products = {
    "Keyboard": 10,
    "Mouse": 25,
    "Monitor": 8,
    "CPU": 18
}

check_inventory(products)
```

```python
# Student Result Processing System

def check_result(marks_list):
    total = 0

    for mark in marks_list:
        total += mark

    average = total / len(marks_list)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

marks = [60, 55, 40, 70, 65]
result = check_result(marks)
print("Result:", result)
```

```python
# Cab Fare Estimator

def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (distance * 12)

    if peak == "yes":
        fare = fare + (fare * 0.25)

    return fare

while True:
    km = float(input("Enter distance in km: "))
    peak_hour = input("Is it peak hour? (yes/no): ").lower()

    total_fare = calculate_fare(km, peak_hour)
    print("Total Fare = ₹", total_fare)

    choice = input("Do you want to calculate again? (yes/no): ").lower()

    if choice != "yes":
        break
```

```python
# Employee Attendance Eligibility Checker

def check_eligibility(attendance_list):
    present_days = 0

    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    print("Attendance Percentage:", percentage, "%")

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
result = check_eligibility(attendance)
print("Status:", result)
```

```python
# Password Strength Checker

def check_password(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False

    # checking minimum length
    if len(password) < 8:
        return "Password too short(Minimum 8 characters required)"

    # checking digit and special character
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    # final validation
    if has_digit and has_special:
        return "Strong Password "
    else:
        return "Weak Password (Must include at least one digit and one special character)"

while True:
    pwd = input("Enter password: ")
    result = check_password(pwd)
    print(result)

    if result.startswith("Strong"):
        break
    else:
        print("Please try again...\n")
```
